```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('1_person_gagu.txt')
df.head()
```

```python
df_by_area = df.set_index(['행정구역별(시도)', '가구주의 연령별'])
df_by_area
```

```python
df_by_area.loc['전국', '합계'].plot()
plt.show()
```

```python
df_national = df_by_area.loc['전국']
df_national
```

```python
s1 = df_national.stack()
s1
```

```python
s2 = s1.rename_axis(['가구 연령별','년도'])
s2
```

```python
s4 = s1.reset_index()
s4.columns = ['가구 연령별','년도','인구']
s4
```

```python
s3 = s2.unstack()
s3
```

```python
t1 = s3.loc[['합계', '65세이상']]
sum_65 = t1.T
sum_65
```

```python
def percent_65(x):
    return (x['65세이상'] / x['합계']) * 100
```

```python
#1인가구 중 65세이상 %
over_65_percent = sum_65.apply(percent_65, axis=1)
over_65_percent
```

```python
over_65_percent.plot()
plt.show()
```

```python
over_65_percent.plot(kind='bar')
plt.show()
```
