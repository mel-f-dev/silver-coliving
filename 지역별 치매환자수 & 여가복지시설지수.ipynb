```python
import pandas as pd
```

```python
df_alzh = pd.read_csv('data/0.치매유병현황_2015-2019(수정).csv', encoding = 'utf-8', header = 11, thousands = ',')
df_alzh.columns=['연도', '지역', '지역 세분', '성별', '연령', '노인인구수', '환자수', '유병률']

df_alz1 = df_alzh[df_alzh['지역'] == df_alzh['지역 세분']]
df_alz1
```

```python
df_alz2 = df_alz1.loc[df_alz1['연령'] =='65세이상']
df_alz2
```

```python
df_alz2.info()
```

```python
df_well = pd.read_csv('data/2. 노인_천명당_노인여가복지시설수_시도_시_군_구__20191202144756.csv', encoding = 'cp949')
df_well = df_well.rename(columns = {'행정구역별(1)':'지역'})
df_well
```

```python
df_well1 = df_well.set_index('지역').T
df_well1
```

```python
df_well2 = df_well1.stack() 
df_well2
```

```python
df_well3 = df_well2.reset_index()
df_well3.columns=['연도', '지역', '노인여가복지시설수(천명당)']
df_well3
```

```python
df_well4 = df_well3.loc[df_well3['지역']!='전국']
df_well4.info()
```

```python
df_well4['연도'] = df_well4['연도'].astype('int')
```

```python
df_well4
```

```python
df_silver = df_alz2.set_index(['연도','지역']).join(df_well4.set_index(['연도','지역']))
```

```python
df_silver['유병률']=df_silver.유병률.str.replace('%','').astype('float64')
```
