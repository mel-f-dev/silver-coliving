```python
import pandas as pd
```

```python
df_ter = pd.read_csv('data/3.전국_도별__논밭별_경지면적_20191202174233.csv', encoding = 'cp949')
df_ter.to_csv('전국_경지면적', encoding = 'utf-8')
```

```python
# 계
df_ter = df_ter.loc[3:]
df_ter
```

```python
df_ter1 = df_ter.loc[df_ter['전답별'] == '계']
df_ter2 = df_ter1.rename(columns = {'시도별':'지역'})
df_ter2
```

```python
df_ter3 = df_ter2.drop(df_ter2.columns[1], axis = 'columns')
df_ter3
```

```python
df_ter4 = df_ter3.set_index('지역').T
df_ter4 = df_ter4.rename(columns = {'제주도':'제주특별자치도'})
df_ter4
```

```python
df_ter5 = df_ter4.stack().reset_index()
df_ter5.columns = ['연도', '지역', '경지면적']
df_ter5
```

```python
df_ter5.to_csv('경지면적', encoding = 'utf-8')
```
