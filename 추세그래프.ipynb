```python
# 치매인구
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_silver = pd.read_csv('지역별_노인여가복지시설수(천명당)')
df_silver
```

```python
df_silver.set_index('지역')
```

```python
seoul = df_silver[df_silver['지역'] == '서울특별시']
seoul
```

```python
seoul.plot(kind = 'scatter', x = '노인여가복지시설수(천명당)', y = '유병률', figsize=(7,7));
```

```python
df_corr = seoul[['유병률', '노인여가복지시설수(천명당)']].corr()
df_corr
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = '노인여가복지시설수(천명당)', y = '유병률', hue = '지역', data= df_silver, s=100, palette = 'dark')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.lmplot(x = '노인여가복지시설수(천명당)', y = '유병률', hue = '지역', data = df_silver)
plt.show()
```

```python
s_senior = pd.read_csv('data/독거노인비율 - 독거노인비율.csv', index_col = 0, header = 1, thousands = ',')
s_senior
```

```python
s_senior1 = s_senior.T
s_senior1
```

```python
s_senior2 = s_senior1.reset_index()
```

```python
s_senior2.rename(columns={'index' : '연도'}, inplace=True)
s_senior2.to_csv('s_senior', encoding = 'utf-8')
s_senior2
```

```python
s_senior3 = s_senior2.set_index('연도')
s_senior3
```

```python
s_senior2['연도'].value_counts().plot(kind='barh')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 7))

# # Load the example car crash dataset
senior_pop = pd.read_csv('s_senior')

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(y="65세 이상 인구(명)", x="연도", data=senior_pop,
            label="65세 이상 인구(명)", color="g")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(y="65세 이상 1인가구수", x="연도", data=senior_pop,
            label="65세 이상 1인가구수", color="g")

# Add a legend and informative axis label
ax.legend(ncol=1, loc="lower right", frameon=True)
ax.set(xlim=(-3, 10), ylim=(0,8000000), ylabel="",
       xlabel="년도별 65세 이상 노인 중 1인가구 비율")
sns.despine(left=True, bottom=True)
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x = s_senior3.index, y = '65세 이상 1인가구수', data = s_senior3)
plt.show()
```

```python
plt.figure(figsize = (20,7))
sns.countplot(x = '연도', hue ='독거노인비율(%)', data = s_senior2)
plt.show()
```
